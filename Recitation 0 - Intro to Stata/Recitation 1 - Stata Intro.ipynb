## Loops

```stata
foreach v of varlist weight rep78 foreign{
    reg price `v'
}
```

```stata
forvalues i = 1/5{
    reg price weight foreign if rep78 == `i'
}
```

```stata
local i = 1
while `i' <= 5{
    reg price weight foreign if rep78 == `i'
    local i = `i' + 1
}
```

```stata
levelsof rep78, local(levels)
foreach i of local levels{
    reg price weight foreign if rep78 == `i'
}
```
